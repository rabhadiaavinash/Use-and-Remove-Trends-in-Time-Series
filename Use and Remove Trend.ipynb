```python
# Difference the Shampoo Sales dataset.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def parser(x):
    return pd.datetime.strptime('190'+x, '%Y-%m')
```

```python
series = pd.read_csv('shampoo_sales.csv', header=0, index_col=0, parse_dates=True,
squeeze=True, date_parser=parser)
```

```python
series
```

```python
plt.plot(series)
plt.show()
```

```python
# shows Trend
```

```python
# Detrend by differencing
```

```python
X = series.values
```

```python
diff = list()
```

```python
for i in range(1, len(X)):
    value = X[i] - X[i - 1]
    diff.append(value)
```

```python
plt.plot(diff)
plt.show()
```

```python
# Trend Removed
```

```python
# Detrend by model fitting
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
X = [i for i in range(0, len(series))]
```

```python
X
```

```python
X = np.reshape(X, (len(X), 1))
```

```python
X
```

```python
y = series.values
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
# calculate trend
trend = model.predict(X)
```

```python
# plot trend
plt.plot(y)
plt.plot(trend)
plt.show()
```

```python
# detrend
detrended = [y[i]-trend[i] for i in range(0, len(series))]
```

```python
# plot detrended
plt.plot(detrended)
plt.show()
```
